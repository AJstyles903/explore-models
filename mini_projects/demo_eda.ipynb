# üìä Summary Statistics ‚Äì `data.describe()`

## üîπ What It Does
The `describe()` function in pandas generates descriptive statistics for a DataFrame or Series.

It summarizes:
- Central tendency
- Dispersion
- Distribution shape

> ‚ö†Ô∏è NaN values are excluded automatically.

---

## üîπ Default Statistics Returned

| Statistic | Meaning |
|------------|----------|
| **count** | Number of non-null observations |
| **mean**  | Average value |
| **std**   | Standard deviation |
| **min**   | Minimum value |
| **25%**   | First quartile (25th percentile) |
| **50%**   | Median (50th percentile) |
| **75%**   | Third quartile (75th percentile) |
| **max**   | Maximum value |

---

## üîπ Code Example

```python
data.describe()
```

---

## üîπ Why It‚Äôs Useful
Provides a quick statistical overview of numerical data during Exploratory Data Analysis (EDA).

### Pandas:
- This is a powerful data manipulation library in Python.
- It provides data structures and functions needed to manipulate structured data,
- including functions for reading and writing data,
- handling missing data,
- merging datasets,
- reshaping data, and more.
- It's built on top of two core Python libraries - Matplotlib for data visualization and NumPy for mathematical operations.

### Matplotlib.pyplot:
- This is a plotting library in Python for 2D graphics.
- It provides a way to visually inspect data.
- The pyplot module of Matplotlib provides a MATLAB-like interface for making plots and charts.
- It has functions for creating a variety of plots including line plots, scatter plots, bar plots, error bars, histograms, pie charts, etc.

### Seaborn:
- This is a statistical data visualization library in Python.
- It's built on top of Matplotlib and closely integrated with Pandas data structures.
- Seaborn helps you explore and understand your data by providing a high-level interface for drawing attractive and informative statistical graphics.
- It offers built-in themes for styling Matplotlib graphics and tools for visualizing univariate and bivariate data,
- comparing different subsets of data, and visualizing linear regression models.

````python
# Age (year)
data.age.describe().round(4)
````

````python
# Visualize the distribution of the target variable
sns.countplot(x ='default', data = data)
plt.show()
````

````python
# count plot on two variables
sns.countplot(x ='gender', hue = "default", data = data)

# Show the plot
plt.show()
````

````python
# Visualize the distribution of all the features
data.hist(figsize=(20,15), rwidth=0.9, bins=20, grid=False)
plt.show()
````

````python
# Check the correlation between the features

# Select only numeric features
numeric_features = data.select_dtypes(include=['number'])

# Check the correlation between the numeric features
corr_matrix = numeric_features.corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
````
