```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
news = fetch_20newsgroups()
news.keys()
```

```python
count_vect = CountVectorizer()
x_train_cv = count_vect.fit_transform(news.data)
print(x_train_cv)
```

```python
tfid = TfidfTransformer()
x_train_tf = tfid.fit_transform(x_train_cv)
print(x_train_tf)
```

```python
multi_nb = MultinomialNB()
multi_nb.fit(x_train_tf,news.target)
```

```python
x_train_cv = count_vect.transform(news.data)
x_train_tf = tfid.transform(x_train_cv)
predict = multi_nb.predict(x_train_tf)
predict
```

```python
accuracy_score(news.target,predict)
```

```python
confusion_matrix(news.target,predict)
```
