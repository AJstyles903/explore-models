```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
```

```python
data = pd.DataFrame(pd.read_csv('./diabetes_dataset.csv'))
data
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.drop('Outcome',axis=1),data.Outcome,test_size=0.2,random_state=42)
```

```python
dec_tr = DecisionTreeClassifier(random_state=75)
dec_tr.fit(x_train,y_train)
```

```python
plot_tree(dec_tr)
```

```python
print(export_text(dec_tr))
```

```python
y_test_pred = dec_tr.predict(x_test)
y_test_pred
```

```python
accuracy_score(y_test,y_test_pred)
```

```python
confusion_matrix(y_test,y_test_pred)
```
