```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.datasets import load_iris
data = load_iris()
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.data, data.target, train_size=0.2, random_state=42)
```

```python
rndClfr = RandomForestClassifier(random_state=100)
rndClfr.fit(x_train,y_train)
y_test_pred = rndClfr.predict(x_test)
```

```python
accuracy_score(y_test,y_test_pred)
```

```python
confusion_matrix(y_test,y_test_pred)
```
