```python
from scipy.stats import norm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.DataFrame(pd.read_csv('../sample_data/sample.csv'))[['V11']]
out_data = data.V11
```

```python
data.describe()
```

```python
data_mean, data_std = out_data.mean(), out_data.std()
cut_off = data_std * 3
lower, upper = data_mean - cut_off, data_mean + cut_off
```

```python
cut_off = data_std * 3
lower, upper = data_mean - cut_off, data_mean + cut_off
```

```python
plt.figure(figsize = (10,5))
plt.hist(data,bins=80,density=True,rwidth=0.9)
rng = np.linspace(out_data.min(), out_data.max(), 100)
plt.plot(rng,norm.pdf(rng, data_mean, data_std),color='red',linewidth=2)
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.histplot(
    data=data,
    x='V11',
    bins=80,
    stat='density',   # IMPORTANT: match scale with PDF
    kde=False         # we will draw KDE separately
)
# KDE line (full control)
sns.kdeplot(
    data=data,
    x='V11',
    color='yellow',
    linewidth=2,
    linestyle='--'
)
plt.axvspan(xmin = lower,xmax= out_data.min(),alpha=0.2, color='red')
plt.axvspan(xmin = upper,xmax= out_data.max(),alpha=0.2, color='red')
plt.show()
```
