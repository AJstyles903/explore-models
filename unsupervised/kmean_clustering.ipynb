```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.DataFrame({
    "x":[78,64,86,86,88,46,28,45,68,64,66,53,82,33,10],
    "y":[58,34,96,56,18,86,78,35,78,94,76,13,42,93,19]
})
data.head()
```

```python
kmeans = KMeans(n_clusters=10,random_state=42)
kmeans.fit(data)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
data["k_mean_label"] = labels
data.head()
```

```python
plt.scatter(data.x,data.y,50,labels)
for i in centroids:
    plt.scatter(i[0],i[1],100,marker="*")
plt.show()
```

```python
inertia = []

for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)

plt.plot(range(1,10), inertia, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()
```
